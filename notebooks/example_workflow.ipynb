```python
import pycomlink as pycml
```

### Load data

```python
cml_id = 'MY4320_2_MY2291_4'
cml = pycml.io.get_cml_data_from_IFU_database(cml_id, '2014-10-10', '2014-12-30')
```

### Plot TX-RX level

```python
cml.plot_txrx(figsize=(16,6), resampling_time='15min');
```

```python
cml.do_wet_dry_classification()
cml.data.wet_nf.plot(figsize=(16,6));
cml.data.wet_fn.plot(figsize=(16,6));
```

```python
cml.do_baseline_determination(method='linear')
cml.calc_A()
cml.data.A_nf.plot(figsize=(16,6))
cml.data.A_fn.plot(figsize=(16,6));
```
